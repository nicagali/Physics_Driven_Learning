# Generating Netwokrs

I generate networks using the package NetworkX [https://networkx.org/]. 

>One of the possible directions of the project is to study the training as a function of **network geometry**. 

For this reason, it is useful to be able to easily create complex geometries or geometries with peculiar features.

For all of them:

- Nodes are divided into three types: *source* nodes (red), *target* nodes (purple) and *hidden* nodes (gray)
- Each node corresponds to a value of voltage
- Each edge corresponds to a value of resistance. All the graphs are initialized with a vlue of R = 50 k$\Omega$.

*Here I list the graphs I consider.*

To generate a graph, use module <font color="orange">graphs</font> and call function <font color="yellow">"name_graph"()</font>. 

To print a graph, use module <font color="orange">plotting_functions</font> and call function <font color="yellow">plot_graph(<font color="purple">"name_graph"</font>)</font>.

### Random Graph (random_graph)

Generate a random graph given # of nodes (<font color="purple">nodes</font>), # of edges (<font color="purple">edges</font>), # of sources (<font color="purple">numb_sources</font>), # of targets (<font color="purple">numb_targets</font>) and input voltages (<font color="purple">volt_sources</font>). The function randomly distributes nodes and establishes connections between them.

To generate the graph call function <font color="yellow">random_graph(<font color="purple">nodes, edges, numb_sources, numb_targets, volt_sources</font>)</font> in module <font color="orange">graphs</font>. To plot, as above.

Here, the graph changes every time you run, so you need to both generate it and plot it.

```python
# Go to parameters module and set the desired values of graph parameters

fig, ax = plt.subplots()
graphs.random_graph(save_data = True)
plotting_functions.plot_graph(ax, fig,'random_graph')
```
